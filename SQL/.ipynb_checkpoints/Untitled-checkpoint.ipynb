```python
import pandas as pd
import pyodbc
import os
```

```python
# List of CSV files and their corresponding table names
csv_files = [
    ('customers.csv', 'customers'),
    ('geolocation.csv', 'geolocation'),
    ('order_items.csv', 'order_items'),
    ('orders.csv', 'orders'),
    ('products.csv', 'products'),
    ('sellers.csv', 'sellers')
]

# Connect to the SQL Server database
conn = pyodbc.connect(
    'Driver={ODBC Driver 18 for SQL Server};'
    'Server=DESKTOP-O4QNJ9S\\SQLEXPRESS;'
    'Database=ecommerce;'
    'UID=sa;'
    'PWD=Login2the*;'
)
cursor = conn.cursor()

cursor.execute("SELECT @@VERSION;")

for row in cursor:
    print(row)

# Folder containing the CSV files
folder_path = r'C:\Users\HP\Desktop\SMIT\SQL Learning\SQL'

def get_sql_type(dtype):
    if pd.api.types.is_integer_dtype(dtype):
        return 'INT'
    elif pd.api.types.is_float_dtype(dtype):
        return 'FLOAT'
    elif pd.api.types.is_bool_dtype(dtype):
        return 'BIT'
    elif pd.api.types.is_datetime64_any_dtype(dtype):
        return 'DATETIME'
    else:
        return 'NVARCHAR(MAX)'

for csv_file, table_name in csv_files:
    file_path = os.path.join(folder_path, csv_file)
    
    df = pd.read_csv(file_path)
    df = df.where(pd.notnull(df), None)

    df.columns = [col.replace(' ', '_').replace('-', '_').replace('.', '_') for col in df.columns]

    # Column definitions
    columns = ', '.join([f'[{col}] {get_sql_type(df[col].dtype)}' for col in df.columns])

    # SQL Server valid CREATE TABLE IF NOT EXISTS method
    create_table_query = f"""
    IF NOT EXISTS (SELECT * FROM sysobjects WHERE name='{table_name}' AND xtype='U')
    CREATE TABLE [{table_name}] ({columns})
    """
    cursor.execute(create_table_query)
    conn.commit()

    # Insert rows using SQL Server placeholder ?
    for _, row in df.iterrows():
        values = tuple(row)
        placeholders = ', '.join(['?'] * len(row))
        insert_query = f"INSERT INTO [{table_name}] ({', '.join([f'[{col}]' for col in df.columns])}) VALUES ({placeholders})"
        cursor.execute(insert_query, values)

    conn.commit()

conn.close()
print("All CSV files imported successfully!")
```
